## Exploratory Data Analysis

### Outline of my approach

My approach to EDA will vary depending on the type of problem I'm looking to solve. For the problem at hand - a binary classification problem in which all of my predictor variables are continuous - I find a good starting point to be splitting up the data by the binary outcomes and looking at how each predictor is distributed for each class. If there is very little overlap in the distribution of variable $x_i$ for malignant samples with the distribution of $x_i$ for benign samples, then that indicates that $x_i$ will be a strong predictor.

**Note:** To avoid clutter, I have chosen to include my preliminary steps at the bottom of this notebook. I start with Seaborn's `sns.pairplot()` function, which gives me a quick way to see distributions of variables, split into categories (along the diagonal). Pairplot is also useful for seeing pairwise correlations between groups of variables, which can be useful for detecting multicollinearity.

After getting a quick visual look at some distributions via pairplots, I use two different types of visualizations to see how distributions of variables differ by diagnosis: **boxplots** and **histograms**. I will explain both below, along with examples of variables. 

In the previous notebook (Data Preparation) I already identified that `'smoothness_mean'` and `compactness_mean` appear to be strong predictors, so I will focus on highlighting other variables in this notebook.

## Boxplots

After identifying some variables where the distributions had relatively less overlap between the 'benign' and 'malignant' samples, I looked at boxplots for each. The boxplot gives a nice view of how much or how little overlap there is between samples. The biggest downside of a boxplot is it can be misleading if the classes are extremely imbalanced. I believe there are enough observations of each label in this dataset to minimize this problem.

```python
fig, ax = plt.subplots(figsize=(16,7), nrows=1, ncols=2)

sns.boxplot(x='diagnosis', y='radius_mean', data=cancer, ax=ax[0])
sns.boxplot(x='diagnosis', y='perimeter_mean', data=cancer, ax=ax[1])

plt.tight_layout();
```

```python
fig, ax = plt.subplots(figsize=(16,7), nrows=1, ncols=2)

sns.boxplot(x='diagnosis', y='concave_points_sd_error', data=cancer, ax=ax[0])
sns.boxplot(x='diagnosis', y='perimeter_sd_error', data=cancer, ax=ax[1])

plt.tight_layout();
```

In the above boxplots, it is clear that the distributions of the variables highlighted differ between benign and malignant samples. To take one particularly striking example, looking at `perimeter_sd_error`, we can see the following:

* The mean perimeter standard error is noticeably higher for malignant observations than for benign ones. In fact, the 25th percentile observation in the malignant sample is above the full box-and-whisker plot for the benign samples (i.e. they would be considered outliers of the benign distribution. I also see that the measurements for malignant samples (for all variables) seem to exhibit significantly greater variance than for the benign samples.

```python
def compare_dist(feature, a=0.05):
    """Function to quickly compare the distribution of a feature by diagnosis.
    Parameters:
    feature - the name of the feature we wish to compare
    a - alpha; significance level for hypothesis test"""
    
    # Print out a histogram of the feature, showing the distribution for each diagnosis
    plt.figure(figsize=(10,7))
    plt.hist(cancer[cancer['diagnosis']==1][feature], bins=20, color='r', alpha=0.5, label='Malignant')
    plt.hist(cancer[cancer['diagnosis']==0][feature], bins=20, color='b', alpha=0.5, label='Benign')
    plt.xlabel(feature)
    plt.legend()
    plt.title(f'Distribution of {feature} by Diagnosis')
    
    # Conduct a quick t-test of the sample means for each diagnosis
    t, p = stats.ttest_ind(cancer[cancer['diagnosis'] == 1][feature],
                       cancer[cancer['diagnosis'] == 0][feature])

    if p < a:
        print(f"""
        With a p-value of {p}, we reject the null that the mean of {feature} 
        is equal between Malignant and Benign samples at the {100*a}% significance level.""")
    else:
        print(f"""
        With a p-value of {p}, we fail to reject the null that 
        the mean of {feature} is equal between Malignant and Benign samples 
        at the {100*a}% significance level.""")
```

```python
compare_dist('radius_mean')
```

```python
compare_dist('smoothness_mean')
```

```python
compare_dist('perimeter_mean')
```

```python
compare_dist('perimeter_sd_error')
```

Above, we can see that the distributions of `radius_mean`, `smoothness_mean`, `perimeter_mean`, and `perimeter_sd_error` share very little overlap, and that their means and medians differ. A two-sample t-test confirms this observation at the 5% significance level (in fact, the extremely small p-values suggest we could have been much more conservative in our choice of signicance level).


Below is one example of a variable that does not appear to be a strong predictor: `compactness_mean`. You can see that the distributions overlap heavily, and we do not reject the null that their means differ.

```python
compare_dist('concavity_mean')
```

Finally, while concavity does not appear to be useful, the standard error of concave points appears to be a very strong indicator of malignant samples - I expect this to be one of the most useful predictors in my classification models.

```python
compare_dist('concave_points_sd_error')
```
